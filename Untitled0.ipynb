```python
import pandas as pd
import json
```

```python
df = pd.read_json('/content/TelecomX_Data.json')
df
```

```python
df.columns
```

```python
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

df_total = pd.concat([df['customerID'],df['Churn'],df_customer, df_phone, df_internet, df_account], axis=1)
df_total_index = df_total.set_index('customerID')
df_total_index.head()
```

```python
df_total_index.info()
```

```python
df_total_Churn = df_total.query('Churn == "Yes"')
df_total_Churn = df_total_Churn.set_index('customerID')
df_total_Churn.head()
```

```python
total_clientes_churn = sum(df_total_Churn['Churn'].value_counts())
total_clientes_Masc = df_total_Churn['gender'].value_counts().get('Male', 0)
total_clientes_Fem = df_total_Churn['gender'].value_counts().get('Female', 0)

print(f'Perncetual cliente genero Feminino {round(((total_clientes_Fem/total_clientes_churn)*100),2)}%.')
print(f'Perncetual cliente genero Masculino {round(((total_clientes_Masc/total_clientes_churn)*100),2)}%.')
```

```python
total_clientes_65 = df_total_Churn['SeniorCitizen'].value_counts().get(1, 0)
total_clientes_65_menos = df_total_Churn['SeniorCitizen'].value_counts().get(0, 0)

print(f'Perncetual cliente Senior {round(((total_clientes_65/total_clientes_churn)*100),2)}%.')
```

```python
import matplotlib.pyplot as plt
```

```python
gender_total = df_total_Churn['gender'].value_counts()
partner_counts_churn = df_total_Churn['Partner'].value_counts()

plt.figure(figsize=(14, 5)) # Ajustar o tamanho para acomodar dois gráficos com mais espaço

plt.subplot(1, 2, 1) # Primeira figura (gênero)
ax1 = gender_total.plot(kind='bar', color=['skyblue', 'lightcoral'])
plt.title('Distribuição de Gênero dos Clientes que Cancelaram')
plt.xlabel('Gênero')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=0)
for container in ax1.containers:
    ax1.bar_label(container)

plt.subplot(1, 2, 2) # Segunda figura (Partner)
ax2 = partner_counts_churn.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title('Quantidade de Clientes que Cancelaram que possuem parceiro')
plt.xlabel('Partner (No: Não, Yes: Sim)')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=0)
for container in ax2.containers:
    ax2.bar_label(container)

plt.tight_layout()
plt.show()

senior_citizen_total_churn = df_total_Churn['SeniorCitizen'].value_counts()

plt.figure(figsize=(8, 5))
ax = senior_citizen_total_churn.plot(kind='bar', color=['lightcoral', 'skyblue'])
plt.title('Quantidade de Clientes que Cancelaram Seniors')
plt.xlabel('Senior Citizen (0: Não, 1: Sim)')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=0)

# Adicionar a quantidade de clientes no topo de cada barra
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()

dependents_total_churn = df_total_Churn['Dependents'].value_counts()

plt.figure(figsize=(8, 5))
ax = dependents_total_churn.plot(kind='bar', color=['lightgreen', 'salmon'])
plt.title('Quantidade de Clientes que Cancelaram possuem  Dependentes')
plt.xlabel('Dependentes (No: Não, Yes: Sim)')
plt.ylabel('Número de Clientes')
plt.xticks(rotation=0)

# Adicionar a quantidade de clientes no topo de cada barra
for container in ax.containers:
    ax.bar_label(container)

plt.tight_layout()
plt.show()
```

```python
columns_to_plot = ['PhoneService', 'MultipleLines', 'InternetService', 'OnlineSecurity',
                   'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV',
                   'StreamingMovies', 'Contract', 'PaperlessBilling', 'PaymentMethod'
                   ]
```

```python
num_columns_to_plot = len(columns_to_plot)
num_cols_per_row = 3
num_rows = (num_columns_to_plot + num_cols_per_row - 1) // num_cols_per_row

print(f"Number of columns to plot: {num_columns_to_plot}")
print(f"Number of columns per row: {num_cols_per_row}")
print(f"Number of rows needed: {num_rows}")
```

```python
fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(18, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    ax = axes[i]

    if col in ['Charges.Monthly', 'Charges.Total', 'tenure']: # Incluindo 'tenure' aqui
        # Para colunas numéricas, usamos histogramas ou box plots
        df_total_Churn[col].hist(ax=ax, bins=20)
    else:
        value_counts = df_total_Churn[col].value_counts()
        value_counts.plot(kind='bar', ax=ax, color=['skyblue', 'lightcoral', 'lightgreen', 'salmon', 'gold'])
        for container in ax.containers:
            ax.bar_label(container)
        ax.tick_params(axis='x', rotation=45) # Rotacionar os rótulos do eixo X

    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

Grupo clientes que cancelaram
Predomonantemente clientes cancelaram contrataram serviços de telefone;
Cerca de metade dos clientes que  cancelaram possuem serviços de multilinhas;
Predominantemente clientes que cancelaram que possuem serviços de internet, possuem fibra optica;
Predominantemente os clientes que cancelaram não realizaram assinaturas adicionais de serviços;
Predominantemente o tipo de contrato é mensal;

Predominantemente realizam seu pagamento por meio de electronic check.

```python
# prompt: criar figura de pizza com percentual de clientes por gasto mensal
# as faixas de gasto mensal deverão iniciar  em até 20 e as proximas faixas com intervalo de 10 em 10
# dados df_Churn_Primeiro_Mes

# Criar as faixas de gasto mensal
bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100, float('inf')]
labels = ['Até 20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100', 'Mais de 100']

# Criar uma nova coluna no DataFrame com as faixas de gasto mensal
df_Churn_Premeiro_Mes['Faixa_Gasto_Mensal'] = pd.cut(df_Churn_Premeiro_Mes['Charges.Monthly'], bins=bins, labels=labels, right=True)

# Contar a quantidade de clientes em cada faixa de gasto mensal
gasto_mensal_counts = df_Churn_Premeiro_Mes['Faixa_Gasto_Mensal'].value_counts().sort_index()

# Criar a figura de pizza
plt.figure(figsize=(10, 8))
plt.pie(gasto_mensal_counts, labels=gasto_mensal_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Percentual de Clientes com Churn no Primeiro Mês por Faixa de Gasto Mensal')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

Grupo de clientes que cancelaram no primeiro mês sob a perpectiva de gasto mensal

A distribuição de gastos mensais demonstra dentro desse grupo a disntinção de três faixas de consumo que podem ser exploradas pela Telecom. São as seguintes faixas:

Faixa 1 gastos entre 20,00 a 30,00 dinheiros;
