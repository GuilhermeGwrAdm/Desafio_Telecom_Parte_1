```python
import pandas as pd
import json
```

```python
df = pd.read_json('/content/TelecomX_Data.json')
df
```

```python
df.columns
```

```python
df_customer = pd.json_normalize(df['customer'])
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])

df_total = pd.concat([df['customerID'],df['Churn'],df_customer, df_phone, df_internet, df_account], axis=1)
df_total_index = df_total.set_index('customerID')
df_total_index.head()
```

```python
df_total_index.info()
```

```python
df_total_Churn = df_total.query('Churn == "Yes"')
df_total_Churn = df_total_Churn.set_index('customerID')
df_total_Churn.head()
```

```python
total_clientes_churn = sum(df_total_Churn['Churn'].value_counts())
total_clientes_Masc = df_total_Churn['gender'].value_counts().get('Male', 0)
total_clientes_Fem = df_total_Churn['gender'].value_counts().get('Female', 0)

print(f'Perncetual cliente genero Feminino {round(((total_clientes_Fem/total_clientes_churn)*100),2)}%.')
print(f'Perncetual cliente genero Masculino {round(((total_clientes_Masc/total_clientes_churn)*100),2)}%.')
```

```python
total_clientes_65 = df_total_Churn['SeniorCitizen'].value_counts().get(1, 0)
total_clientes_65_menos = df_total_Churn['SeniorCitizen'].value_counts().get(0, 0)

print(f'Perncetual cliente Senior {round(((total_clientes_65/total_clientes_churn)*100),2)}%.')
```

```python
fig, axes = plt.subplots(num_rows, num_cols_per_row, figsize=(18, num_rows * 5))
axes = axes.flatten()

for i, col in enumerate(columns_to_plot):
    ax = axes[i]

    if col in ['Charges.Monthly', 'Charges.Total', 'tenure']: # Incluindo 'tenure' aqui
        # Para colunas numÃ©ricas, usamos histogramas ou box plots
        df_total_Churn[col].hist(ax=ax, bins=20)
    else:
        value_counts = df_total_Churn[col].value_counts()
        value_counts.plot(kind='bar', ax=ax, color=['skyblue', 'lightcoral', 'lightgreen', 'salmon', 'gold'])
        for container in ax.containers:
            ax.bar_label(container)
        ax.tick_params(axis='x', rotation=45) # Rotacionar os rÃ³tulos do eixo X

    ax.set_title(col)
    ax.set_xlabel(col)
    ax.set_ylabel('Count')

# Remove any unused subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python
# prompt: criar figura de pizza com percentual de clientes por gasto mensal
# as faixas de gasto mensal deverÃ£o iniciar  em atÃ© 20 e as proximas faixas com intervalo de 10 em 10
# dados df_Churn_Primeiro_Mes

# Criar as faixas de gasto mensal
bins = [0, 20, 30, 40, 50, 60, 70, 80, 90, 100, float('inf')]
labels = ['AtÃ© 20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80', '81-90', '91-100', 'Mais de 100']

# Criar uma nova coluna no DataFrame com as faixas de gasto mensal
df_Churn_Premeiro_Mes['Faixa_Gasto_Mensal'] = pd.cut(df_Churn_Premeiro_Mes['Charges.Monthly'], bins=bins, labels=labels, right=True)

# Contar a quantidade de clientes em cada faixa de gasto mensal
gasto_mensal_counts = df_Churn_Premeiro_Mes['Faixa_Gasto_Mensal'].value_counts().sort_index()

# Criar a figura de pizza
plt.figure(figsize=(10, 8))
plt.pie(gasto_mensal_counts, labels=gasto_mensal_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Percentual de Clientes com Churn no Primeiro MÃªs por Faixa de Gasto Mensal')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```
